Estratégia de tratamento de dados

As colunas com valores ausentes foram analisados conforme a sua semântica.
Variáveis que representam a inexistência de uma característica(ex: piscina, cerca, lareira) terão seus valores ausentes preenchidos com "None".
Variáveis numéricas essenciais terão tratamento estatístico apropriado.


```python
cols_none = [
    "PoolQC",
    "MiscFeature",
    "Alley",
    "Fence",
    "FireplaceQu",
    "GarageType",
    "GarageFinish",
    "GarageQual"
]
```

```python
for col in cols_none :
    train_df[col] = train_df[col].fillna("None")
```

```python
train_df[cols_none].isnull().sum()
```

Após o tratamento, validei a ausência de valores nulos para garantir consistência de dados.

```python
train_df["MasVnrType"] = train_df["MasVnrType"].fillna("None")
```

```python
train_df["MasVnrType"].isnull().sum()
```

```python
train_df[["LotFrontage", "MasVnrArea"]].isnull().sum()
```

```python
lot_frontage_median = train_df["LotFrontage"].median()
train_df["LotFrontage"] = train_df["LotFrontage"].fillna(lot_frontage_median)
```

```python
train_df["MasVnrArea"] = train_df["MasVnrArea"].fillna(0)
```

```python
train_df[["LotFrontage", "MasVnrArea"]].isnull().sum()
```

```python
Electrical_mode = train_df["Electrical"].mode()[0]
train_df["Electrical"] = train_df["Electrical"].fillna(Electrical_mode)
```

```python
train_df["GarageYrBlt"] = train_df["GarageYrBlt"].fillna(0)
```

```python
train_df.isnull().sum().sort_values(ascending=False).head()
```

```python
cols_none = [
    "GarageCond",
    "BsmtExposure",
    "BsmtFinType1",
    "BsmtFinType2",
    "BsmtCond"
]

for col in cols_none:
    train_df[col] = train_df[col].fillna("None")
```

```python
train_df.isnull().sum().sort_values(ascending=False)
```

```python
train_df["BsmtQual"] = train_df["BsmtQual"].fillna("None")
```

```python
train_df["BsmtQual"].isnull().sum()
```
